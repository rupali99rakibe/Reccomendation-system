```python
import pandas as pd
import numpy as np
```

```python
df_credit = pd.read_csv("D:\\DATA_SETS\\tmdb_5000_credits.csv")
df_credit.head()
```

```python
df_movie = pd.read_csv("D:\DATA_SETS\\tmdb_5000_movies.csv")
df_movie.head(2)
```

```python
df_credit.shape
```

```python
df_movie.shape
```

```python
df_concat = pd.concat([df_credit,df_movie],axis=1)
df_concat.head()
```

```python
df = df_concat.drop(['homepage','movie_id','title','status','production_countries'],axis=1)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
v = df['vote_count']
R = df['vote_average']
C = df['vote_average'].mean()
m = df['vote_count'].quantile(0.70)
```

```python
df['Weighted_average'] = ((R*v)+(C*m))/(v+m)
```

```python
df.head(2)
```

```python
df_sorted = df.sort_values('Weighted_average',ascending=False)
df_sorted[['original_title','vote_count','vote_average','Weighted_average','popularity']].head(15)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
weight_average = df.sort_values('Weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=weight_average['Weighted_average'].head(10) , y=weight_average['original_title'].head(10))
plt.xlim(4,10)
plt.xlabel("Weighted_Average_Score",weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.show()
```

```python
weight_average = df.sort_values('Weighted_average',ascending=False)
plt.figure(figsize=(12,6))
axis1 = sns.barplot(x=weight_average['popularity'].head(10) , y=weight_average['original_title'].head(10))
plt.xlabel("Score_Of_Popularity",weight='bold')
plt.ylabel('Movie Title',weight='bold')
plt.title("Most Popular By Votes",weight='bold')
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling = MinMaxScaler()
```

```python
df2 = df_sorted[['Weighted_average','popularity']]
df2.head()
```

```python
df2 = scaling.fit_transform(df_sorted[['Weighted_average','popularity']])
df2_normalized = pd.DataFrame(df2,columns=['Weighted_average','popularity'])
df2_normalized.head()
```

```python
df[['normalized_weight_average','normalized_popularity']] = df2_normalized
```

```python
df.head(3)
```

```python
df['score'] = df["normalized_weight_average"]*0.5+ df["normalized_popularity"]*0.5
scored_df = df.sort_values(['score'],ascending=False)
scored_df = df[['original_title','normalized_weight_average','normalized_popularity','score']]
```
